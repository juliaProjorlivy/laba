```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x_center = 4.78
x_min_left = [5.64, 6.95, 7.26, 7.48, 7.65, 7.84, 8.1]
x_max_left = [5.4, 5.83, 6.8, 7.34, 7.57, 7.73, 7.93]
x_min_right = [4.61, 3.9, 2.63, 2.37, 2.11, 1.91, 1.73]
x_max_right = [3.14, 2.74, 2.48, 2.24, 2.01, 1.83, .166]
ms = [i for i in range(1, 8)]

k = 0.1
x_min_left = [i *  k for i in x_min_left]
x_max_left =  [i *  k for i in x_max_left]
x_min_right = [i *  k for i in x_min_right]
x_max_righ =  [i *  k for i in x_max_right]
lamb = 546 * 10**(-6) #mm
```

```python
x1 = [((x_min_left[i] - x_min_right[i]))**2 / 4 for i in range(len(x_min_left))]
y1 = [(2 * m - 1) * lamb / 2 for m in ms]
fig, ax = plt.subplots()


#plt.legend()
plt.scatter(x1, y1, marker='o', color='blue')
#plt.errorbar(x, y, yerr=dy, xerr=dx, ecolor='k', linestyle='')

model1 = np.poly1d(np.polyfit(x1[2:], y1[2:], 1))
print(model1)

polyline = np.linspace(min(x1), max(x1))
plt.plot(polyline, model1(polyline), color = 'green')

print(model1.c[0])
plt.grid()
plt.xlabel('$r^2$, усл.ед^2')
plt.ylabel('$(1 -1/2)\\lambda$, порядок')
#plt.show()
plt.savefig('tex/images/plot.png', bbox_inches='tight')
```
