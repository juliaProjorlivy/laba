```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

# Данные из таблицы
m = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y1 = np.array([0.25, 0.77, 1.6, 2.16, 2.91, 3.59, 4.22, 5.24])
y2 = np.array([0, 0.46, 1.08, 1.76, 2.5, 3.22, 3.96, 4.58])

plt.scatter(m, y2, marker='o', color='blue')
plt.scatter(m, y1, marker='o', color='red')
plt.legend(["Светлые колца k = 0.70", "Темные кольца k = 0.68"])

polyline = np.linspace(min(m), max(m))
# Построение графика
#plt.figure(figsize=(10, 6))
model2 = np.poly1d(np.polyfit(m, y2, 1))
model1 = np.poly1d(np.polyfit(m, y1, 1))
plt.plot(polyline, model1(polyline), color = 'red')
plt.plot(polyline, model2(polyline), color = 'blue')

print(model1)
print(model2)

# Настройка графика
#plt.title('Зависимость $r_m^2$ и $(r_m\')^2$ от номера $m$')
plt.grid()
plt.ylabel('$r^2$, усл.ед^2')
plt.xlabel('$m$, порядок')
#plt.legend()

plt.savefig('tex/images/plot1.png', bbox_inches='tight')
```
