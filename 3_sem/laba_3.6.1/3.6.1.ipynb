```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Data

```python
an = [282.6 , 274 , 196.9 , 179.8 , 145.6 , 84.83]
ns = [1, 3, 9, 10, 12, 15]
print([round(x / an[0], 3) for x in an])
```

```python
tau = 0.05
T = 1
c1 = math.sin(np.pi * ns[0] * tau / T)/(np.pi * ns[0])
print([math.sin(np.pi * ns[i] * tau / T)/(c1 * np.pi * ns[i]) for i in range(len(ns))])
```

```python
ts = [20, 40, 60, 80, 100, 120, 140, 160, 180, 200]
nus = [46, 25, 17, 12, 10, 8, 7, 5, 5, 5]
```

```python
x = [1 / (0.001 * i) for i in ts]
y = nus
```

```python
fig, ax = plt.subplots()

plt.scatter(x, y, marker='o', color='cornflowerblue')

model1 = np.poly1d(np.polyfit(x, y, 1))
print(model1)

polyline = np.linspace(min(x), max(x))
plt.plot(polyline, model1(polyline), color = 'green', label="k = {: .2f}$\pm 0.01$".format(round(model1.c[0], 2)))

plt.legend()
#ax.set_title('График зависимости $ln(U)(t)$')
plt.errorbar(x, y, yerr=2, ecolor='k', linestyle='')
#ax.plot(x, y, 'r-')
plt.grid()

plt.ylabel('$\\Delta \\nu, кГц$')
plt.xlabel('$1/\\tau, мс^{-1}$')
k = round(model1.c[0], 2)
#plt.show()
plt.savefig('tex/plot1.png', bbox_inches='tight')
```

```python
Ts = [0.6, 1, 1.4, 1.8, 2.2, 2.6, 3, 3.4, 3.8, 4.2, 4.6, 5]
dnu = [1.668, 1.000, 0.723, 0.55, 0.457, 0.383, 0.334, 0.295, 0.264, 0.238, 0.217, 0.200]
```

```python
x = [1 / i for i in Ts]
y = dnu
```

```python
fig, ax = plt.subplots()

plt.scatter(x, y, marker='o', color='cornflowerblue')

model1 = np.poly1d(np.polyfit(x, y, 1))
print(model1)

polyline = np.linspace(min(x), max(x))
plt.plot(polyline, model1(polyline), color = 'green', label="k = {: .2f}$\pm 0.009$".format(round(model1.c[0], 2)))

plt.legend()
#ax.set_title('График зависимости $ln(U)(t)$')
plt.errorbar(x, y, yerr=0.002, ecolor='k', linestyle='')
#ax.plot(x, y, 'r-')
plt.grid()

plt.ylabel('$\\delta \\nu, кГц$')
plt.xlabel('$1/T, мс^{-1}$')
k = round(model1.c[0], 2)
#plt.show()
plt.savefig('tex/plot2.png', bbox_inches='tight')
```
