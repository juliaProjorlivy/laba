```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = 0.1
eL = 0.002
T = 69 * 10**(-6)
dT = 0.1 * 10**(-6)
C0 = T**2 / (L * (2 * np.pi)**2)
eC0 = ((2 * eL)**2 + (dT / T)**2)**0.5
```

```python
print(C0, eC0, C0 * eC0)
```

```python
Ts = [x * 10 ** (-6) for x in [69, 94, 132, 158, 168, 188, 200]]
Cs = [x * 10 ** (-6) + C0 for x in [0, 0.001, 0.003, 0.005, 0.006, 0.008, 0.009]]
```

```python
y = [np.log(i) for i in Ts[1:]]
x = [np.log(i) for i in Cs[1:]]
ek = (eC0**2 + (dT / T)**2)**0.5
```

```python
fig, ax = plt.subplots()
plt.scatter(x, y, marker='o', color='cornflowerblue')

model1 = np.poly1d(np.polyfit(x, y, 1))
print(model1)

k = round(model1.c[0], 3)
polyline = np.linspace(min(x), max(x))
plt.plot(polyline, model1(polyline), color = 'green', label="k = {: .3f}$\pm {: .3f}$".format(k, ek * k))

plt.legend()
plt.grid()

plt.ylabel('$ln(T), мкс$')
plt.xlabel('$ln(С), мкФ$')

#plt.show()
plt.savefig('tex/images/plot1.png', bbox_inches='tight')
```
